# Abstract

Our study aimed to determine which factors: sleep quality, diet, physical activity, social support, academic workload, financial stress, use of counseling service, relationship status, and family history would most strongly correlate to higher depression scores among students aged 18-25 across varying majors. We created multiple visualizations, utilized a random forest classifier, and fit an OLS model to perform an analysis on these factors. Our analysis revealed that semester credit load (academic workload) emerged as the most significant predictor of higher depression scores and highlighted the interconnected nature of lifestyle and social factors in shaping mental health outcomes.

# Research Question

For males and females, ages 18-25, in various college major (Engineering, Medical, Law, etc), which of the factors are most highly correlated with a higher depression score:

- **Sleep Quality Level** (Good = Wake Up Feeling Refreshed, Average = Sleep Enough, Poor = Sleep Very Little)
- **Diet Quality** (Good = 3+ Meals a Day, Average = 2 Meals a Day, Poor = 0-1 Meal)
- **Physical Activity Level** (High = Go to the Gym, Moderate = Walking/Slow Jog, Low = Don't Exercise)
- **Social Support** (High = Feels that there is always someone to talk to when needed, Moderate, Low)
- **Amount of School Work** (Semester Credit Load)
- **Financial Pressure** (0 (Living Comfortably) - 5 (Very Stress About Money))
- **Usage of Counseling Services** (Frequently = Goes Weekly, Occasionally = Goes Monthly, Never = Haven't Been)
- **Relationship Status** (Single, Married, In A Relationship)
- **Family History of Depression** (Yes/No)

Additionally, are students of certain majors, gender, university, or age groups more prone to be affected by depression?

## Background and Prior Work


The prevalence of depression among college students has become a significant public health concern, exacerbated by increasing academic pressures, financial strain, and social challenges. College students, particularly those aged 18-25, are in a transitional life stage characterized by greater independence, shifts in social networks, and heightened academic expectations, all of which can contribute to mental health challenges. Studies indicate that depression is widespread within this demographic, with research finding that 53% of American college students reported experiencing depression since starting college, and 9% contemplated suicide (Furr et al., 2021).<a name="ref-2"></a>[<sup>2</sup>](#cite-2) Similarly, it is observed that over a quarter of Swiss university students exhibited depressive symptoms during the COVID-19 pandemic, significantly exceeding both general population and pre-pandemic levels (Volken et al., 2021).<a name="ref-3"></a>[<sup>3</sup>](#cite-3)

Prior research has examined various predictors of depression among college students, often focusing on isolated factors such as socioeconomic status, academic workload, or lifestyle habits. Liu's team and Liu conducted a literature review categorizing non-pathological depression contributors into biological, psychological, academic, and lifestyle factors, emphasizing the role of socioeconomic status, family support, and health behaviors in mitigating depressive symptoms (Liu et al., 2022).<a name="ref-1"></a>[<sup>1</sup>](#cite-1) Additionally, studies have identified certain student subgroups, such as art students and those in Institutes of Technology, as being at higher risk due to field-specific stressors and socioeconomic pressures (McLafferty et al., 2022).<a name="ref-4"></a>[<sup>4</sup>](#cite-4) Researchers have also explored the role of big data and machine learning in depression prediction, utilizing behavioral data from fitness trackers, smartphones, and social networks to assess mental health trends.

However, while previous studies have provided valuable insights into the factors influencing depression, there remains a gap in understanding how multiple lifestyle, social, and academic variables interact to contribute to depression across different academic disciplines. Our research questions seeks to address this gap by investigating which factors—such as sleep quality, diet, physical activity, social support, financial stress, academic workload, counseling service utilization, relationship status, and family history of depression—are most strongly correlated with depression scores among male and female college students from various academic disciplines (e.g., Engineering, Medical, Law, etc.). By examining a more comprehensive set of lifestyle and academic factors within a diverse student population, we aim to provide a more nuanced understanding of the determinants of depression and whether students of certain majors, genders, universities, or age groups are disproportionately affected.

```python
# Drop Columns According to the factors we will be analyzing specifically
df = df.drop(columns = ['CGPA', 'Stress_Level', 'Anxiety_Score', 
                        'Substance_Use', 'Chronic_Illness', 
                        'Extracurricular_Involvement', 'Residence_Type'])
df.head()
```

## Predictive Analysis

To analyze every feature's importance, while worrying about overfitting and noise of the 12 factors (input features) we will be using, we decided to use the random forest classifier. Since a random forest model combines the output of multiple randomized decision trees to reach one best result for that training data using the mode voting method, it has the following benefits:

- High Accuracy
- Reduced Overfitting
- Robust to Noise
- High Dimensional Dataset Compatibility
- Non-Parametric Nature (No need for a certain type of data distribution)

Additionally, since it is utilizing multiple decision trees, it also come with the feature important evaluation that we are looking for in this project, much similar to that of a decision trees.

From the EDA, we noticed that our data does not contain any outlier, and that most of our distribution are approximately uniform, therefore there is no need to transform our quantitative variables. As for our categorical variables, we use the sklearn's `OneHotEncoder` to tranform it into numerical values of 0's and 1's. Then we dropped the first column of the one-hot encoded feature to avoid redundancy and remove false linear dependency. The idea is if we recieve a row of all 0's, then it means that feature belongs to the first category of column that was removed, else if the rows have a 1 anywhere in the row, the category is under the column where the 1 is. Additionally, noticing how most of the factor's distribution of depression score is approximately uniform across many depression score at the same time, it will be hard for the classifier to determine which score is correct, meaning if we create a multi-class classifier that predicts the depression score from 0 to 5, it is most likely that the classifier will make equal prediction to 0 to 3, all the low scores, and 4 to 5, all the high scores. Therefore, we believe that it is best to create a classifier to determine if the depression score is high (4-5) or low (0-3). As our original research question is about which factors is highly correlated to a higher depression score, this split between high and low, will still work for our question.

Therefore, using the 80:20 Train:Test split ratio, we trained and test our data. In this test, we will only create one forest as our goal is to not fine tune the hyperparameter to create a higher accuracy model, but to rank the feature's importance.

```python
# Create matrix of features
X = df.drop(['Depression_Score'], axis = 1)
# Create a series of depression score of 1 and 0, where 1 correspond
# to high depression score and 0 to low depression score
y = (df['Depression_Score'] >= 4).astype(int)

# Split the training and testing dataset by 80 and 20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Create a pipeline with transformer and classifier 
# for high and low depression score
rand_forest_clf = make_pipeline(
    make_column_transformer((OneHotEncoder(drop = 'first'), categorical_col),
                            remainder = 'passthrough'),
    # Forest with 100 decision tree
    RandomForestClassifier(n_estimators = 100)
)

# Fit and transform the training dataset
rand_forest_clf.fit(X_train, y_train)

# Print accuracy of the classification prediction after predicting
# for both training set and test set
print('Training:', rand_forest_clf.score(X_train, y_train), 
      '\nTesting:', rand_forest_clf.score(X_test, y_test))
```

The `feature_importances_` in the random forest classifier return numbers corresponding to the importance of each individual columns in making the decision, but since we one-hot encoded the categorical factors and created multiple columns for the same factor, we will get multiple number. Noticing how the values in feature importances altogether adds to 1, we decided to sum the individual category of the same factor to get a single value of importance for each factor.

```python
features_importance = {}
for feature, value in zip(rand_forest_clf[:-1].get_feature_names_out(), 
                          rand_forest_clf[1].feature_importances_):
    feature = feature.split('__')
    if feature[0] == 'onehotencoder':
        feature = '_'.join(feature[1].split('_')[:-1])
        if feature in features_importance:
            features_importance[feature] += value
        else:
            features_importance[feature] = value
    else:
        features_importance[feature[1]] = value

features_importance = pd.DataFrame(features_importance.values(), 
                                   index = features_importance.keys(),
                                   columns = ['Importance Value'])
features_importance = features_importance.sort_values('Importance Value', 
                                                      ascending = False)
features_importance
```

```python
(sns.barplot(x = features_importance.index, 
            y = 'Importance Value', 
            data = features_importance)
    .tick_params(axis = 'x', 
                 rotation = 90))
plt.title('Importance of Each Factors to Depression Evaluation')
plt.xlabel('Factors')
plt.ylabel('Proportion of Relative Importance')
```

The Random Forest Classifier provided insights into the relative importance of these factors influencing students' depression scores. The most influential factors identified were Semester Credit Load (importance ~ 0.17), Age (importance ~ 0.14), and Course (importance = 0.13). To see if these importance were of any significance, we decided to do a significance analysis of the top factor.

## Significance Analysis

To assess the significance of the features we found as important in the predictive analysis, we decided to do some significance test on our top factors with the following rejection significance level in mind.

**Significance Level:** 0.05

```python
credit_model = smf.ols(formula = 'Depression_Score ~ Semester_Credit_Load', data = df).fit()
credit_model.summary()
```

Since the p-value, 0.006, is less than our significance value, 0.05, we say that student's semester credit load has a significant affect on student's depression score.

```python
age_model = smf.ols(formula = 'Depression_Score ~ Age', data = df).fit()
age_model.summary()
```

Since the p-value, 0.920, is greater than our signifcance value, 0.05, we have insuffient evidence that our importance score for age is a significant important feature towards depression score.

Since we reject the signficance of one of our top features of high importance according to our random forest classifier, we believe that the rest of the other features with a much lower importance score, will also be insignificant, therefore, we stopped our significance analysis here.
