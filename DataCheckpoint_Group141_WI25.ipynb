The prevalence of depression among college students has become a significant public health concern, exacerbated by increasing academic pressures, financial strain, and social challenges. College students, particularly those aged 18-25, are in a transitional life stage characterized by greater independence, shifts in social networks, and heightened academic expectations, all of which can contribute to mental health challenges. Studies indicate that depression is widespread within this demographic, with research finding that 53% of American college students reported experiencing depression since starting college, and 9% contemplated suicide (Furr et al., 2021).<a name="ref-2"></a>[<sup>2</sup>](#cite-2) Similarly, it is observed that over a quarter of Swiss university students exhibited depressive symptoms during the COVID-19 pandemic, significantly exceeding both general population and pre-pandemic levels (Volken et al., 2021).<a name="ref-3"></a>[<sup>3</sup>](#cite-3)

Prior research has examined various predictors of depression among college students, often focusing on isolated factors such as socioeconomic status, academic workload, or lifestyle habits. Liu's team and Liu conducted a literature review categorizing non-pathological depression contributors into biological, psychological, academic, and lifestyle factors, emphasizing the role of socioeconomic status, family support, and health behaviors in mitigating depressive symptoms (Liu et al., 2022).<a name="ref-1"></a>[<sup>1</sup>](#cite-1) Additionally, studies have identified certain student subgroups, such as art students and those in Institutes of Technology, as being at higher risk due to field-specific stressors and socioeconomic pressures (McLafferty et al., 2022).<a name="ref-4"></a>[<sup>4</sup>](#cite-4) Researchers have also explored the role of big data and machine learning in depression prediction, utilizing behavioral data from fitness trackers, smartphones, and social networks to assess mental health trends.

However, while previous studies have provided valuable insights into the factors influencing depression, there remains a gap in understanding how multiple lifestyle, social, and academic variables interact to contribute to depression across different academic disciplines. Our research questions seeks to address this gap by investigating which factors—such as sleep quality, diet, physical activity, social support, financial stress, academic workload, counseling service utilization, relationship status, and family history of depression—are most strongly correlated with depression scores among male and female college students from various academic disciplines (e.g., Engineering, Medical, Law, etc.). By examining a more comprehensive set of lifestyle and academic factors within a diverse student population, we aim to provide a more nuanced understanding of the determinants of depression and whether students of certain majors, genders, universities, or age groups are disproportionately affected.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the dataframe
df = pd.read_csv('students_mental_health_survey.csv')
df.head()
```

```python
#Check the size of the dataframe (row by column)
df.shape
```

```python
# Drop Columns According to the factors we will be analyzing specifically
df = df.drop(columns = ['CGPA', 'Stress_Level', 'Anxiety_Score', 
                        'Substance_Use', 'Chronic_Illness', 
                        'Extracurricular_Involvement', 'Residence_Type'])
df.head()
```

```python
# Filter rows according to the age group of higher education student 
# we are specifically looking at
df = df[(df['Age'] >= 18) & (df['Age'] <= 25)]
df.head()
```

```python
# Check if there are any null values to handle
df.isna().sum().sum()
```

```python
# Get shape of new dataframe
df.shape
```

```python
# Understand the datatype of the dataframe we will work with later on
df.dtypes
```

```python
# Get the unique value for each column in the dataframe
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in column '{column}':")
    print(unique_values, '\n')
```

```python
# Plot distribution of integer values
fig_1, ax_1 = plt.subplots(2, 2)
row = 0
col_n = 0
for col in ['Age', 'Depression_Score', 'Financial_Stress', 'Semester_Credit_Load']:
    # Account for exclusive endpoint in np.arange and inclusive endpoint in bins on the last bin
    bins = np.arange(min(df[col]), max(df[col]) + 2)
    df[col].hist(bins = bins, ax = ax_1[row][col_n])
    ax_1[row][col_n].set_xlabel(col)
    ax_1[row][col_n].set_ylabel('Frequency')
    row += 1
    if row == 2:
        row = 0
        col_n += 1

fig_1.tight_layout()
fig_1.show()
```

```python
# Plot distribution of categorical variables
fig_2, ax_2 = plt.subplots(3, 3)
row = 0
col_n = 0
for col in df.columns:
    if df[col].dtype == object:
        sns.countplot(x = col, data = df, ax = ax_2[row][col_n])
        ax_2[row][col_n].tick_params(axis = 'x', rotation = 90)
        row += 1
        if row == 3:
            row = 0
            col_n += 1
fig_2.set_size_inches(12,12)
fig_2.tight_layout()
fig_2.show()
```
